```python
base_url = 'https://mediabiasfactcheck.com/{}'
biases = ['left', 'leftcenter', 'center', 'right-center', 'right']

for bias in biases:  #iterate through the nav bars 
    mbfc_links = []
    driver.get(base_url.format(bias))
    time.sleep(7)
    print("done sleeping")
    table = driver.find_element_by_xpath('//*[@id="mbfc-table"]')
    print("table found")
    sources = table.find_elements_by_tag_name('td')
    # print(f"sources: {sources}")
    for source in sources: #each source is a row in the table
        # print(f"source {source.text}")
        try: 
            # ele = source.find_element_by_xpath("./span/a")
            ele = source.find_element_by_tag_name("a")
            name = ele.text
            name_split = name.split(' (')
            source_link = name_split[1] if len(name_split) == 2 else ""
            source_link = source_link.strip(")")
            page_link = ele.get_attribute("href")
            mbfc_links.append((name_split[0], source_link, page_link))
            print(f"{name_split[0]},  {source_link}, {page_link}")
            
        except Exception as e: #there are some empty rows in the table and this may cause an error, skip them
            print(e)
            continue
    with open(f"mbfc_{bias}.json", "w", encoding="utf-8") as outfile:
        json.dump(mbfc_links, outfile)
```

```python
links = ['https://mediabiasfactcheck.com/left', 'https://mediabiasfactcheck.com/leftcenter', 
         'https://mediabiasfactcheck.com/center', 'https://mediabiasfactcheck.com/right-center', 
         'https://mediabiasfactcheck.com/right']
sourcedict = {}
for link in links[:1]:  #iterate through the nav bars   
    driver.get(link)
    time.sleep(7)
    print("done sleeping")
    table = driver.find_element_by_xpath('//*[@id="mbfc-table"]')
    print("table found")
    sources = table.find_elements_by_tag_name('td')
    print(f"sources: {sources}")
    for source in sources[:3]: #each source is a row in the table
        print(f"source {source.text}")
        try: 
            # ele = source.find_element_by_xpath("./span/a")
            ele = source.find_element_by_tag_name("a")
            name = ele.text
            sourcedict[name] = {}
            print(name)
            
            ActionChains(driver).key_down(Keys.CONTROL).click(ele).key_up(Keys.CONTROL).perform()
            
            oldtab = driver.window_handles[0]
            newtab = driver.window_handles[1]
            
            driver.switch_to.window(newtab)
            time.sleep(4)
            sourcedict[name]['bias'] = link[link.find('m/')+2:]            
                        
            report = driver.find_element_by_xpath('/html/body/div[2]/div/div/article/div/p[2]')
                                                
            for i,item in enumerate(report.find_elements_by_tag_name('strong')): #all bolded items
                text = str(item.text)
                    
                print('**** text:',text)
                if len(text)<5: print('\ttoo short:',text) #pass on country acronyms like USA & bias labels like HIGH and LEFT
                    
                elif 'CREDIBILITY' in text:  sourcedict[name]['credibility'] = text
                elif 'organization' in text.lower() or 'website' in text.lower() or 'TV' in text.lower(): sourcedict[name]['source type'] = text
                elif '/' in text and 'organization' not in text.lower(): #text[-1].isnumeric(): 
                    if 'USA' in text: sourcedict[name]['country'] = 'USA' 
                    else: sourcedict[name]['country'] = text[:text.find(' ')]
                    sourcedict[name]['world press freedom rank'] = text[text.find(' ')+1:]
                    
                elif 'Traffic' in text: sourcedict[name]['traffic'] = text
                elif text[-1].isupper() and not text[-1].isnumeric(): sourcedict[name]['factual reporting']=text
                elif len(text.split()) ==1: sourcedict[name]['country'] = text
                else: sourcedict[name]['source type'] = text
                #print('\t',title,':',item.text)
                #sourcedict[name][title] = item.text
            
            # JS: adding source link
            sle = driver.find_elements_by_tag_name('p')
            for e in sle:
                if e.text[:5] == "Source": sourcedict[name]['source_link'] = e.find_element_by_tag_name('a').getAttribute('href')
            print(sourcedict[name])    
            driver.close()
            driver.switch_to.window(oldtab)
        except Exception as e: #there are some empty rows in the table and this may cause an error, skip them
            if 'no such element' not in str(e): print(e)
            
            if sourcedict[name].get('report') != None: #if no source was created, name will be last source, and already has info
                print('entering if branch')
                continue
            #driver.close()
            #driver.switch_to.window(oldtab)
            break
```

```python
def format(x):
    if type(x) == str: 
        parts = x.split('\n')
        if '\n' not in x: return x
        if 'Bias Rating' in parts[0]: parts.remove(parts[0])
        return parts
    return ''

def get_attribute(x, attribute):
    for x in x:
        if attribute in x: return x
    else: return ''

def fix_country(x):
    if '(' in x: return x[x.find('(')+1:x.find(')')] 
```

```python
df['report'] = df['report'].apply(format)
attributes = ['Media Type','Factual Reporting', 'Country:', 'World Press', 'MBFC Credibility Rating:','Traffic/Popularity']
for a in attributes:
    df[a] = df['report'].apply(lambda x: get_attribute(x, a).strip(a+':'))

```
